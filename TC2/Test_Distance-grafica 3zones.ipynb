```python
# Se cargan las librerias 
# esta es una version de carlos saenz 
import numpy as np
from matplotlib import pyplot as plt

# importing "cmath" for complex number operations 
import cmath 
```

Se define la función para pasar de polares a cartesianos

```python
def zones_plot(Z1L, zone1, zone2, zone3):# que es ZL1 ? la magnitud en que unidades ?
    r, theta = cmath.polar(Z1L)
    fig, ax = plt.subplots(1, 1, figsize =(6, 6))
    fig.suptitle('Diagrama de impedancia para relé de característica Mho', y=0.95, fontsize=16)
    
    r1=r*zone1
    r2=r*zone2
    r3=r*zone3
    #theta = np.linspace(0, 2*np.pi, 360)
    #x1 = r*np.cos(theta)
    #x2 = r*np.sin(theta)
    #x3 = r*np.sin(theta)
    #ax.plot(x1,x2,'k--')
    
    # Creando el punto de origen para los vectores
    x_pos = [0, 0,0] 
    y_pos = [0, 0,0]
    
    x_Z1L = Z1L.real
    y_Z1L = Z1L.imag
    
    ax.set(xlim=(-r1, r1*2), ylim = (-r1, r1*2)) #que hace esta monda 

    #se crean los circulos 
    
    a_circle = plt.Circle((x_Z1L*zone1/2, y_Z1L*zone1/2), r1/2, fill = False, linestyle= '--', edgecolor = 'r')
    ax.add_artist(a_circle)
    
    a_circle2 = plt.Circle((x_Z1L*zone2/2, y_Z1L*zone2/2), r2/2, fill = False, linestyle= '--', edgecolor = 'g')
    ax.add_artist(a_circle2)
    
    a_circle3 = plt.Circle((x_Z1L*zone3/2, y_Z1L*zone3/2), r3/2, fill = False, linestyle= '--', edgecolor = 'b')
    ax.add_artist(a_circle3)
    
    #que hace quiver ??
    
    ax.quiver(x_pos, y_pos, x_Z1L, y_Z1L, angles='xy', scale_units = 'xy', scale=1, color=['k'], label='Line Impedance')
    
    
    plt.text(-r/10, r*zone1+r/10, r'$Z_1$', {'color': 'red', 'fontsize': 12}, rotation=0, horizontalalignment='center', verticalalignment='top',
         multialignment='center')
    
    plt.text(-r/10, r*zone2+r/10, r'$Z_2$', {'color': 'green', 'fontsize': 12}, rotation=0, horizontalalignment='center', verticalalignment='top',
         multialignment='center')
    
    plt.text(-r/10, r*zone3+r/10, r'$Z_3$', {'color': 'blue', 'fontsize': 12}, rotation=0, horizontalalignment='center', verticalalignment='top', multialignment='center')
        
    #Create broken lines
    x=np.linspace(start=-r1,stop=r1*2,num=30)
    plt.scatter(x,[0]*len(x), marker='_',color='k')
    plt.scatter([0]*len(x), x, marker='|',color='k')
    
    plt.xlabel('R')
    plt.ylabel('X')
    plt.legend()
```

```python
# Inicializando la data
L = 45 # Line Total Length 
R1L = 0.1 #[ohm/mille]
X1L = 0.7 #[ohm/mille]
R0L = 0.1 #[ohm/mille]
X0L = 0.7 #[ohm/mille

Z1L = complex(R1L*L, X1L*L)
Z0L = complex(R0L*L, X0L*L)
zone1= 0.8
zone2=1.2
zone3=1.4
zones_plot(Z1L, zone1, zone2, zone3)
```
