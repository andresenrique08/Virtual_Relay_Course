```python
import numpy as np
from matplotlib import pyplot as plt 

# Reading voltaje and currents
voltages_bus1 = np.empty(([len(time),3]))
currents_bus1 = np.empty(([len(time),3]))

voltages_bus2 = np.empty(([len(time),3]))
currents_bus2 = np.empty(([len(time),3]))

col=0
for i in ['a','b','c']:
    voltages_bus1[:,col] = mat['vVt1'+i].ravel()
    currents_bus1[:,col] = mat['iCt1'+i].ravel()
    voltages_bus2[:,col] = mat['vVt2'+i].ravel()
    currents_bus2[:,col] = mat['iCt2'+i].ravel()
    col=col+1
            
f, axarr = plt.subplots(2, 2, figsize =(16, 10))
f.suptitle('Lectura del archivo .mat', y=0.97, fontsize=16)

axarr[0,0].plot(time, voltages_bus1[:,0], 'b-', label='Fase A')
axarr[0,0].plot(time, voltages_bus1[:,1], 'r-', label='Fase B')
axarr[0,0].plot(time, voltages_bus1[:,2], 'g-', label='Fase C')
axarr[0,0].set_xlabel('Tiempo [seg]')
axarr[0,0].set_ylabel('Voltaje [V]')
axarr[0,0].grid()
#axarr[0,0].legend()
axarr[0,0].set_title('Voltaje de bus 1')

axarr[0,1].plot(time, currents_bus1[:,0], 'b-', label='Fase A')
axarr[0,1].plot(time, currents_bus1[:,1], 'r-', label='Fase B')
axarr[0,1].plot(time, currents_bus1[:,2], 'g-', label='Fase C')
axarr[0,1].set_xlabel('Tiempo [seg]')
axarr[0,1].set_ylabel('Corriente [A]')
axarr[0,1].grid()
#axarr[0,1].legend()
axarr[0,1].set_title('Corriente de bus 1')

axarr[1,0].plot(time, voltages_bus2[:,0], 'b-', label='Fase A')
axarr[1,0].plot(time, voltages_bus2[:,1], 'r-', label='Fase B')
axarr[1,0].plot(time, voltages_bus2[:,2], 'g-', label='Fase C')
axarr[1,0].set_xlabel('Tiempo [seg]')
axarr[1,0].set_ylabel('Voltaje [V]')
axarr[1,0].grid()
#axarr[1,0].legend()
axarr[1,0].set_title('Voltaje de bus 1')

axarr[1,1].plot(time, currents_bus2[:,0], 'b-', label='Fase A')
axarr[1,1].plot(time, currents_bus2[:,1], 'r-', label='Fase B')
axarr[1,1].plot(time, currents_bus2[:,2], 'g-', label='Fase C')
axarr[1,1].set_xlabel('Tiempo [seg]')
axarr[1,1].set_ylabel('Corriente [A]')
axarr[1,1].grid()
#axarr[1,1].legend()
axarr[1,1].set_title('Corriente de bus 1')

labels = ['Fase A', 'Fase B', 'Fase C']
f.legend(axarr[1,1].get_lines(), labels, ncol=3, bbox_to_anchor=(0.54, 0.87))
```
